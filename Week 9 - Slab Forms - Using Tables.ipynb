In the Formwork text many of the same tables we used for wall forms are the ones we use for slab forms.  Starting with the Plyform span limits found in Table 7-2.  Again, Table 7-2 has values for 1/2" to 1", wet, Class I Plyform, with both $C_D = 1.0$ and $C_D = 1.25$, in either the strong or weak orientation.  The other sheathing tables are for other support conditions.

Tables 7-5, 7-6, and 7-7 are similar to the sheathing tables; they are for different support conditions, and both duration factors.  This time we will use Table 7-5 for the Joist and Stringer designs.  Keep in mind that these beam tables are arranged such that the .1 version contains values found using $C_D = 1.0$, and the .2 version contains values using $C_D= 1.25$.  Both the .1 and .2 versions are for No.2 Douglas Fir-Larch lumber and the versions .3 and .4 are based on using No.1/No.2 grade Spruce Pine Fir.

There are no wales in slabs so we will not be using Tables 7-8 through 7-10.

Table 7-11 contains allowable axial loads of a variety of species, sizes, and effective lengths.  The values in this table are only for a $C_D=1.0$.  This is not unreasonable since you are much more likely to reuse shores which means their capacity would have to be calculated using a $C_D=1.0$.  As I mentioned previously there are six obvious typos in the table.   Everywhere the table should indicate a value for a 10 ft unsupported length, (between 9 ft and 11 ft) the table reads 19 ft; you should strike that and write 10 next to it.

Since the procedure for slab forms is very similar to wall forms we will move straight on to an example.

Design formwork for a concrete floor slab based on the following information:

* Thickness of concrete slab = 10”
* Unbraced length for 4x6 shores = 14 ft
* Concrete unit weight = 150 lb/ft$^3$
* Deflection limited to L/360 but not greater than 1/16”
* All form lumber will be S4S No.2 Douglas Fir-Larch
* Assume dry conditions for lumber and wet conditions for plywood
* Decking for the slab will be 3/4” Class I Plyfrom
* Load duration factor of 7 days for everything except the shores.
* Dead Load for formwork material = 5.5 psf and Live Load = 50 psf

Since the problem statement did not indicate otherwise, the load duration factor $C_D = 1.25$.  This means,we are using the left hand side of Table 7-2.  The concrete pressure is over $175 psf$ in the left hand column, so we have to use $200 psf$.

For $3/4"$ Class I Plyform, subjected to a $200 psf$ pressure load, using a 7 day load duration factor, orientating the Plyform in the strong direction (grain perpendicular to the supports), and supported over three or more spans, the maximum support spacing is $19"$. 

Ass a reminder, for the purposes of preparing you for the [AIC exam](http://www.professionalconstructor.org/exam-info), use $19"$ on center.
